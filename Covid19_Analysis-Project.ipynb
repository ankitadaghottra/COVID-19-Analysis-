```python
df1 = pd.read_csv(r'C:\Users\Dell\Documents\DataSets\COVID-19-Data-Analyisis-master\covid_19_data.csv', parse_dates = ['Last Update'])
```

```python
df1.rename(columns={'ObservationDate': 'Date', 'Country/Region': 'Country'}, inplace=True)
```

```python
df1.query('Country=="India"')
```

```python
df1.groupby('Date').sum()
```

```python
ConfirmeCases = df1.groupby('Date').sum()['Confirmed'].reset_index()
ConfirmedDeaths = df1.groupby('Date').sum()['Deaths'].reset_index()
ConfirmedRecovery = df1.groupby('Date').sum()['Recovered'].reset_index()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = ConfirmeCases['Date'], y = ConfirmeCases['Confirmed'], mode = 'lines+markers', name =  'Confirmed Cases',line = dict(color = 'green', width = 2)))
fig.add_trace(go.Scatter(x = ConfirmedDeaths['Date'], y = ConfirmedDeaths['Deaths'], mode = 'lines+markers', name =  'Confirmed Deaths',line = dict(color = 'blue', width = 2)))
fig.add_trace(go.Scatter(x = ConfirmedRecovery['Date'], y = ConfirmedRecovery['Recovered'], mode = 'lines+markers', name =  'Confirmed Recovery',line = dict(color = 'red', width = 2)))
```
